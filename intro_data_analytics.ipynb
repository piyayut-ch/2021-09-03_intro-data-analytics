# Data Wrangling

## Filter

```r
gapminder %>%
  filter(country == "Thailand")
```

```r
# เลือกเฉพาะข้อมูลของประเทศ ญี่ปุ่น
gapminder %>%
  filter(country == "")
```

## Arrange

```r
gapminder %>%
  arrange(continent)
```

```r
gapminder %>%
  arrange(country, desc(year))
```

```r
# ในปี 2007 เรียงลำดับตามจำนวนประชากร
gapminder %>%
  filter() %>%
  arrange()
```

## Select

```r
gapminder %>%
  select(country, year, pop)
```

```r
# เลือกข้อมูลชื่อประเทศ และชื่อทวีป
gapminder %>%
  select()
```

## Mutate

```r
gapminder %>%
  mutate(pop_m = pop/1000000) %>%
  filter(country == "Thailand")
```

```r
# แปลงข้อมูล gdpPercap เป็น บาท โดยสมมติว่า 1 US dollar = 30
gapminder %>%
  mutate(gdpPercap_baht = )
```

## Summarize

```r
gapminder %>%
  group_by(continent, year) %>%
  summarize(pop = sum(pop)/1000000)
```

```r
# หาอายุคาดหวังเฉลี่ย รายทวีป function ที่คือ mean
gapminder %>%
  group_by(continent, year) %>%
  summarize(lifeExp = )
```

## Join

```r
gapminder %>%
  left_join(country_codes, by = c("country" = "country"))
```

## Pivot Wider

```r
gapminder %>%
  pivot_wider(
    id_cols = "country",
    names_from = "year",
    values_from = "lifeExp")
```

```r
# ตารางประชากรรายประเทศรายปี หน่วย ล้านคน
gapminder %>%
  mutate() %>%
  pivot_wider(
    id_cols = "",
    names_from = "",
    values_from = "")
```

## Pivot Longer

```r
gapminder %>%
  pivot_longer(
    cols = c("lifeExp", "pop", "gdpPercap"),
    names_to = "variable",
    values_to = "values")
```

# Data Visualization

## Bar Chart

```r
ggplot(
  data = gapminder %>% 
    filter(
      year == 2007, 
      continent == "Americas"),
  aes(
    x = lifeExp, 
    y = reorder(country, lifeExp))) + 
  geom_col()
```

```r
# กราฟแท่งประชากรของทวีปเอเชีย ในปี 2002
ggplot(
  data = gapminder %>% 
    filter(
      year == , 
      continent == ""),
  aes(
    x = , 
    y = reorder(country, ))) + 
  geom_col()
```

## Line Chart

```r
ggplot(
  data = gapminder %>% 
    filter(country %in% c("Thailand", "Japan")),
  aes(
    x = year, 
    y = lifeExp, 
    color = country)) + 
  geom_line()
```

```r
# line chart รายได้ของไทยเทียบกับเวียดนาม
ggplot(
  data = gapminder %>% 
    filter(country %in% c("Thailand", "")),
  aes(
    x = , 
    y = , 
    color = country)) + 
  geom_line()
```

## Histogram

```r
ggplot(
  data = gapminder %>% filter(year == 2007),
  aes(x = lifeExp)) + 
  geom_histogram(
    bins = 20, 
    fill = "navyblue", 
    color = "white")
```

```r
# histogram ประชากร ของปี 1957 
ggplot(
  data = gapminder %>% filter(year == ),
  aes(x = )) + 
  geom_histogram(
    bins = 20, 
    fill = "navyblue", 
    color = "white")
```

## Treemap

```r
ggplot(gapminder %>% filter(year == 2007), 
       aes(area = pop,
           fill = continent, 
           subgroup = continent,
           label = country
          )) +
  geom_treemap(color = "white") +
  geom_treemap_subgroup_border(color = "grey40") +
  geom_treemap_text(
    colour = "grey30", 
    place = "topleft", 
    reflow = TRUE) +
  theme(legend.position = "none")
```

```r
# treemap ประชากร ของปี 1952
ggplot(gapminder %>% filter(year == ), 
       aes(area = pop,
           fill = continent, 
           subgroup = continent,
           label = country
          )) +
  geom_treemap(color = "white") +
  geom_treemap_subgroup_border(color = "grey40") +
  geom_treemap_text(
    colour = "grey30", 
    place = "topleft", 
    reflow = TRUE) +
  theme(legend.position = "none")
```

## Scatter Plot

```r
ggplot(data = gapminder %>% filter(year == 2007), 
       aes(x = gdpPercap, 
           y = lifeExp, 
           size = pop, 
           col = continent)) + 
  geom_point(alpha = 0.5) +
  scale_x_log10() +
  scale_size(range = c(1, 15))
```

```r
# scatter plot ของปี 1987
ggplot(data = gapminder %>% filter(year == ), 
       aes(x = gdpPercap, 
           y = lifeExp, 
           size = pop, 
           col = continent)) + 
  geom_point(alpha = 0.5) +
  scale_x_log10() +
  scale_size(range = c(1, 15))
```
